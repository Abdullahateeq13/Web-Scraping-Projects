# All SaaS Companies

```python
# Websites URL = "https://golden.com/query/list-of-software-as-a-service-companies-BMY"
```

```python
# Importing Important Libraries

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import StaleElementReferenceException
from selenium.common.exceptions import ElementClickInterceptedException
import time
import pandas as pd

# Lists
company_names = []
company_descriptions = []
company_websites = []

# Driver Initialize
driver = webdriver.Chrome()  # Update this with the path to your chromedriver executable

# Wait 
wait = WebDriverWait(driver, 30)

# Getting URL
driver.get("https://golden.com/query/list-of-software-as-a-service-companies-BMY")

# Loop for finding elements on each page
# We are finding only 24 pages from the website as there are 100 pages available, which will be more time taken....
for i in range(1, 25):
    print(i)   # Printing the page number...
    
    # Select 100 items per page
    select = Select(driver.find_element(By.XPATH, '//div[@class="PageSizePicker"]//select'))
    select.select_by_value("100")
    time.sleep(2)
    
    
    # Apply while loop here, because some pages have problems and multiple errors happening.
    while True:
        try:
            # Company Names
            names = driver.find_elements(By.XPATH, '//div[@class="css-1ci65ym"]//a[@class="TopicLink__link legacyLink"]')
            
            break
        except StaleElementReferenceException:
            pass
    
    time.sleep(2)
    
    while True:
        try:
            # Descriptions
            descriptions = driver.find_elements(By.XPATH, '//p[@class="Editor__text"]')
             
            break
        except StaleElementReferenceException:
            pass
    
    time.sleep(2)
    
    while True:
        try:
            # Websites
            websites = driver.find_elements(By.XPATH, '//div[1]/div[@class="css-1ci65ym"][1]/a[@class="css-s8dpi2"][1][1]')

            break
        except StaleElementReferenceException:
            pass
     
    # Check if lengths are the same before appending the data.
    
    if len(names) == len(descriptions) == len(websites):
        company_names.extend([name.text for name in names])
        company_descriptions.extend([desc.text for desc in descriptions])
        company_websites.extend([web.text for web in websites])

    # Find the button for the current page and click on it
    button = driver.find_element(By.XPATH, "//button[@class='css-1vv7e4v']//div[@class='css-vurnku' and text()='" + str(i+1) + "']")

    
    # Click the button
    while True:
        try:
            button.click()
            break
        except ElementClickInterceptedException:
            time.sleep(1)
    
    time.sleep(5)

# Close the browser
driver.quit()
    
# Assingning lists into dictionaries
columns = {'Names':company_names,
           'Descriptions': company_descriptions,
           'Websites': company_websites
          }

# Saving columns into DataFrame
df = pd.DataFrame(columns, index=None)

# Show first 5 lines of df
df.head()

# Shape of the df
df.shape

```
